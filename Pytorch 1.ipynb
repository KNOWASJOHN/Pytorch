#Pytorch Reproducability

(Trying to take random out of random)
How neural network works
1. take random numbers
2. train
3. update parameters

To reduce randomness in Pytorch and neural networks

```python
import torch
t1 = torch.rand(3, 4)
t2 = torch.rand(3, 4)
print(t1)
print(t2)
print(t1 == t2)
```

# lets make some random but reproducable seed

```python
RANDOM = 40
torch.manual_seed(RANDOM)
t2 = torch.rand(3, 4)
print(t1)
print(t2)
print(t1 == t2)
```
